## **Linear Regression Model** ğŸ¦Š
-----------------------

### **Data Preprocessing ğŸº**

#### All data is in numeric form, no need for any preprocessing âœ”ï¸

### **Recognizing Features & Targets** ğŸ¦

```python
# Splitting features & targets ğŸŒ±
# Feature Data 
x = df.iloc[:,0:3]
# Target Data 
y = df.iloc[:,-1] 
```

### **Data Train & Test Split** ğŸ‚

```python
# Splitting data into train and test datasets ğŸ‚
X_train, X_test, Y_train, Y_test = train_test_split(x,y,test_size=0.2,random_state=2)
```

### **Model Training & Predition ğŸ™**

```python
# Training data on Linear Regression Algorithm ğŸ›ï¸
lr = LinearRegression()
# Model Training ğŸ“
lr.fit(X_train, Y_train)
```

#### **Model Evaluation** ğŸ—¿

```python
# Model prediction ğŸ’
pred = lr.predict(X_test)
# R Square value to check credibility of the model ğŸš§
r2Value = r2_score(Y_test,pred)
print("r2 score value : ",r2Value)
```

### **Building a Predictive System** ğŸ‹

```python
class Pred:
    
    def __init__(self):
        self.sales = None
        
    def model(self, tv, radio, newspaper):
        
        data = np.array([tv, radio, newspaper])
        data = data.reshape(1,-1)
        sales = lr.predict(data)
        
        return int(sales)
    
if __name__ == '__main__':
        obj = Pred()
        t = input("\n Please enter cost of advertisement on TV : ")
        r = input("\n Please enter cost of advertisement on radio : ")
        n = input("\n Please enter cost of advertisement on newspaper : ")
        r = obj.model(t,r,n)
        print("\n Sales Predicted are : ", int(r))
```
